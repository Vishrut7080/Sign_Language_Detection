```python
import os
```

```python
#os.listdir('Sign_Language_Dataset')
'''
nc: 5
'''
train_path= os.path.join('Sign_Language_Dataset','dataset_split','images','train')
test_path= os.path.join('Sign_Language_Dataset','dataset_split','images','test')
val_path= os.path.join('Sign_Language_Dataset','dataset_split','images','val')
train_labels_path= os.path.join('Sign_Language_Dataset','dataset_split','labels','train')
val_labels_path= os.path.join('Sign_Language_Dataset','dataset_split','labels','val')
test_labels_path= os.path.join('Sign_Language_Dataset','dataset_split','labels','test')
```

```python
names= ['yes', 'no', 'hello', 'iloveyou', 'thankyou']
```

```python
from tensorflow.keras.utils import load_img, img_to_array
import numpy as np
```

```python
def load_images(path):
    images=[]
    for file in os.listdir(path):
        file_path=os.path.join(path,file)
        i=load_img(file_path, target_size=(224,224))
        i=img_to_array(i)
        images.append(i)
    return np.array(images)
```

```python
train_images=load_images(train_path)
```

```python
#loading and normalizing validation images
val_images=load_images(val_path)
val_images/=255
```

```python
#loading and normalizing testing images
test_images=load_images(test_path)
test_images/=255
```

```python
def load_labels(path):
    labels=[]
    for file in os.listdir(path):
        file_path=os.path.join(path, file)
        with open(file_path,'r') as f:
            contents=f.read()
            id=contents.split()[0]
            labels.append(int(id))
    return np.array(labels)
```

```python
train_labels=load_labels(train_labels_path)
val_labels=load_labels(val_labels_path)
test_labels=load_labels(test_labels_path)
```

```python
print (len(train_labels), len(val_labels), len(test_labels))
```

# Model Development

```python
from tensorflow.keras import models, layers
```

```python
model=models.Sequential([
    # Data augmentation
    layers.RandomFlip("horizontal", input_shape=(224,224,3)),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
    
    layers.Conv2D(filters=32, kernel_size=(3,3), activation="relu", input_shape=(224, 224, 3)),
    layers.MaxPooling2D(pool_size=(2,2)),

    layers.Conv2D(filters=64, kernel_size=(4,4), activation="relu"),
    layers.MaxPooling2D(pool_size=(2,2)),
   
    layers.Flatten(),
    layers.Dense(64, activation="relu"),
    layers.Dense(len(names), activation="softmax")
])
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
history=model.fit(
    train_images, train_labels,
    epochs=125,
    batch_size=32,
    validation_data=(val_images, val_labels)
)
```

```python
import matplotlib.pyplot as plt

# Plot training & validation accuracy
plt.figure(figsize=(12, 5))

# Accuracy
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy', marker='o')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', marker='o')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

# Loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss', marker='o')
plt.plot(history.history['val_loss'], label='Validation Loss', marker='o')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
